## 3. Data Consistency Checks

### 3.1 Data Types

```python
# Print of names of columns
df.columns
```

```python
# Print of number of rows and columns
df.info()
```

```python
# Apply advanced python function that checks df for mixed type columns
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# Change data type of 'PUB_AGENCY_UNIT' column to string
df['PUB_AGENCY_UNIT'] = df['PUB_AGENCY_UNIT'].astype('str')
```
