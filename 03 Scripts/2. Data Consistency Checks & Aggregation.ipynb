## 3. Data Consistency Checks

### 3.1 Data Types

```python
# Print of names of columns
df.columns
```

```python
# Print of number of rows and columns
df.info()
```

```python
# Apply advanced python function that checks df for mixed type columns
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# Change data type of 'PUB_AGENCY_UNIT' column to string
df['PUB_AGENCY_UNIT'] = df['PUB_AGENCY_UNIT'].astype('str')
```

```python
# Check data type of 'PUB_AGENCY_UNIT' column
df['PUB_AGENCY_UNIT'].dtype
```

```python
# Change data type of 'OFFENDER_RACE' column to string
df['OFFENDER_RACE'] = df['OFFENDER_RACE'].astype('str')
```

```python
# Check data type of 'OFFENDER_RACE' column
df['OFFENDER_RACE'].dtype
```

```python
# Change data type of 'OFFENDER_ETHNICITY' column to string
df['OFFENDER_ETHNICITY'] = df['OFFENDER_ETHNICITY'].astype('str')
```

```python
# Check data type of 'OFFENDER_ETHNICITY' column
df['OFFENDER_ETHNICITY'].dtype
```

```python
# Apply advanced python function that checks df for mixed type columns
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# Print of first five rows of df
df.head()
```

```python
# replace 'nan' values with 'NaN' for consistency
df_1 = df.replace('nan', np.NaN)
```

```python
# Change data type of 'INCIDENT_ID' column to string as doesn't represent number with statistical importance
df_1['INCIDENT_ID'] = df_1['INCIDENT_ID'].astype('str')
```

### 3.2 Missing Data

```python
# Find missing values in df
df_1.isnull().sum()
```

```python
# Create new dataframe dropping columns with too many missing values
df_drop = df_1.drop(columns = ['PUB_AGENCY_UNIT','ADULT_VICTIM_COUNT','JUVENILE_VICTIM_COUNT','ADULT_OFFENDER_COUNT','JUVENILE_OFFENDER_COUNT'])
```

```python
# Check shape of dataframe
df_drop.shape
```

```python
# Check head of dataframe
df_drop.head()
```

### 3.3 Duplicate Data

```python
# Search for duplicated rows in df_drop and create a subset that contains only these rows
df_dups = df_drop[df_drop.duplicated()]
```

```python
# Visualise df_dups subset that contains duplicated rows
df_dups
```

```python
# There are no duplicated rows
```

### 3.4 Aggregate Data

```python
# Find count of crimes by state
df_drop.groupby('STATE_NAME').agg({'DATA_YEAR':['count']})
```

```python
# Find count of crimes by year
df_drop.groupby('DATA_YEAR').agg({'STATE_NAME':['count']})
```
